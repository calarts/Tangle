```
from PIL import Image, ImageDraw, ImageFont
from math import sqrt
import requests
import matplotlib.pyplot as plt
# from matplotlib.pyplot import imshow, figure

plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['figure.edgecolor'] = 'white'

# use a truetype font
fnt = ImageFont.truetype('IBMPlexSans-Medium.ttf', 21)
bfnt = ImageFont.truetype('slkscr.ttf', 36)

%matplotlib inline
```

```
pandaUrl   = 'https://www.evernote.com/l/ADOw2tfbp7lE8p-f7yxcK74ghHwVuvR4v0IB/image.png'
jaguarUrl  = 'https://www.evernote.com/l/ADNsAD0jWDtGvYq4I85KOUQXz9HDwCN4M-YB/image.png'
gorillaUrl = 'https://www.evernote.com/l/ADOdBwKmLZ5G6q8ClR2-Ru50tWj_v6cgofkB/image.png'
rhinoUrl   = 'https://www.evernote.com/l/ADNAollR6kFPzrtD0ETXuZ9ycRSraoijuCYB/image.png'
leopardUrl = 'https://www.evernote.com/l/ADMNhBAFfAdL9r0WZRstcqQ9XqIcDuN82b0B/image.png'
orangUrl   = 'https://www.evernote.com/l/ADOG4tgj-WpHVaH-F5O4z4lMJzdSQDzHl10B/image.png'

panda = Image.open(requests.get(pandaUrl, stream=True).raw)
jaguar = Image.open(requests.get(jaguarUrl, stream=True).raw)
gorilla = Image.open(requests.get(gorillaUrl, stream=True).raw)
rhino = Image.open(requests.get(rhinoUrl, stream=True).raw)
orang = Image.open(requests.get(orangUrl, stream=True).raw)
```

```
fig = plt.figure(figsize=(10,10), facecolor='white', edgecolor='white')

# At least count (2014 census) there were 1,864 giant pandas living in the wild.
cnt = 1864
msg = str(cnt) + "\nGiant Pandas."

# Use a square image and you can use the square root
xy = int(sqrt(cnt))

# Resize it down to match the "extinct" resolution, 
# then back up to the original 400x400 pixels.
# use Nearest Neighbor interpolation to retain the Aliasing blocks
extinct = panda.resize((xy, xy)).resize((400,400),resample=Image.NEAREST)
draw  = ImageDraw.Draw(extinct)
draw.text( (10,310), msg, fill='#FFFFFF', font=bfnt)
extinct.save('extinct_pandas.png')

plt.imshow(extinct)
```
